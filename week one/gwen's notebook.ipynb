```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#import data
df = pd.read_csv('Levels_Fyi_Salary_Data.csv')
```

```python
#get rid of outliers in total yearly compensation
df = df[df['totalyearlycompensation'] < 40000]

#get rid of outliters in years of expierence 
df = df[df['yearsofexperience'] < 50]
```

```python
#check headings
df.head()
```

```python
#see what data is missing
df.isna().sum()
```

```python
df.yearsofexperience.hist()
plt.xlabel('years of expierence')
plt.ylabel('frequency')
plt.title('distribution of years of expierence')
```

```python
df.yearsatcompany.hist()
plt.xlabel('years at current company')
plt.ylabel('frequency')
plt.title('distribution of years at current company')
```

```python
df.totalyearlycompensation.hist()
plt.xlabel('total yearly compensation')
plt.ylabel('frequency')
plt.title('distribution of total yearly compensation')
```

```python
plt.figure(figsize=(15, 15))
sns.scatterplot(data = df, x='yearsofexperience',y='totalyearlycompensation')
```

```python
plt.figure(figsize=(15, 15))
sns.scatterplot(data = df, x='yearsatcompany',y='totalyearlycompensation')
```
